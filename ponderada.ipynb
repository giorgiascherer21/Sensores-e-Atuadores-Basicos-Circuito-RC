## Introdução

Este notebook apresenta a análise dos dados coletados na atividade do circuito RC com Arduino UNO. Após a montagem e execução do experimento no protoboard, os valores de tensão, apresentados no Monitor Serial, foram registrados pelo Arduino e exportados para este ambiente, permitindo visualizar graficamente o comportamento de carga e descarga do capacitor.

O objetivo central é interpretar os resultados gerando gráficos correspondentes, possibilitando observar como o resistor influencia o tempo de resposta do circuito. 

```python
# Importando a biblioteca matplotlib.pyplot
import matplotlib.pyplot as plt
```

```python

# ============================
# DADOS 
# ============================
dados = [
    (0, 5.00, 0.00),
    (401, 4.80, 0.20),
    (803, 4.61, 0.39),
    (1204, 4.43, 0.57),
    (1605, 4.25, 0.75),
    (2006, 4.07, 0.93),
    (2408, 3.89, 1.11),
    (2809, 3.70, 1.30),
    (3210, 3.52, 1.48),
    (3611, 3.33, 1.67),
    (4012, 3.15, 1.85),
    (4414, 2.96, 2.04),
    (4815, 2.78, 2.22),
    (5216, 2.59, 2.41),
    (5617, 2.41, 2.59),
    (6018, 2.22, 2.78),
    (6419, 2.04, 2.96),
    (6820, 1.85, 3.15),
    (7221, 1.67, 3.33),
    (7622, 1.48, 3.52),
    (8023, 1.30, 3.70),
    (8424, 1.11, 3.89),
    (8825, 0.93, 4.07),
    (9226, 0.75, 4.25),
    (9627, 0.57, 4.43),
    (10028, 0.39, 4.61),
    (10429, 0.20, 4.80),
    (10830, 0.00, 5.00)
]
```

```python
# ============================
# Separando os dados em listas individuais
# ============================
tempo = [item[0] for item in dados]
tensao_c = [item[1] for item in dados]  # Tensão no capacitor
tensao_r = [item[2] for item in dados]  # Tensão no resistor
```

```python
# ============================
# Gráfico 1 - Carga do capacitor
# ============================
plt.figure(figsize=(10, 5))
plt.plot(tempo, tensao_c, color='blue', marker='o')
plt.title('Carga do Capacitor (Circuito RC)')
plt.xlabel('Tempo (ms)')
plt.ylabel('Tensão no Capacitor (V)')
plt.grid(True)
plt.show()
```

### Análise dos Resultados - Gráfico 1
Neste primeiro gráfico, observamos que a tensão no capacitor inicia em aproximadamente 5V e diminui gradualmente até 0V ao longo do tempo. Esse comportamento indica o processo de descarga do capacitor, em que a energia armazenada é liberada através do resistor.


Embora a curva apresentada seja praticamente linear, em um circuito RC real essa variação tende a ser exponencial, pois a taxa de descarga é mais intensa no início e desacelera conforme o capacitor se aproxima de zero.

Portanto, podemos concluir que o gráfico confirma que o capacitor libera sua energia gradualmente, demonstrando o processo de descarga em um circuito RC e a relação direta entre a tensão no capacitor e o tempo.

```python
# Gráfico 2 - Descarga do resistor
# ============================
plt.figure(figsize=(10, 5))
plt.plot(tempo, tensao_r, color='red', marker='o')
plt.title('Descarga no Resistor (Circuito RC)')
plt.xlabel('Tempo (ms)')
plt.ylabel('Tensão no Resistor (V)')
plt.grid(True)
plt.show()
```

### Análise dos Resultados - Gráfico 2
No segundo gráfico, a tensão no resistor parte de 0V e cresce até cerca de 5V, comportamento inverso ao do capacitor. Isso ocorre pois, à medida que o capacitor se descarrega, a corrente flui pelo resistor, gerando uma diferença de potencial crescente até que o sistema atinja o equilíbrio.


```python
# Gráfico 3 - Comparação lado a lado
# ============================
plt.figure(figsize=(10, 6))
plt.plot(tempo, tensao_c, label='Tensão no Capacitor (Carga)', color='blue')
plt.plot(tempo, tensao_r, label='Tensão no Resistor (Descarga)', color='red')
plt.title('Comparação - Carga x Descarga (Circuito RC)')
plt.xlabel('Tempo (ms)')
plt.ylabel('Tensão (V)')
plt.legend()
plt.grid(True)
plt.show()
```

### Análise dos Resultados - Gráfico 3
No terceiro gráfico, as duas curvas: 
- Azul (tensão no capacitor) 
- Vermelha: (tensão no resistor) 
são apresentadas simultaneamente. 

A curva azul decresce de 5V para 0V, enquanto a vermelha cresce na mesma proporção, cruzando-se por volta de 2,5V e 5.400 ms.

Esse ponto de cruzamento representa o momento em que a energia está igualmente distribuída entre os dois componentes do circuito. 

O comportamento entre as curvas reforça o princípio fundamental de um circuito RC, que no caso afirma que a soma das tensões nos dois elementos se mantém próxima ao valor total da fonte, evidenciando a transferência de energia entre capacitor e resistor ao longo do tempo.
